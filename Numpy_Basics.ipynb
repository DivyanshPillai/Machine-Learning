#Mathematical Operation on an Array

```python
x = np.random.randint(10, 20, (3, 3))
print(x)
y = np.random.randint(20, 30, (3, 3))
print(y)
```

**One way to do multiplication operations in an array**

```python
print("Addition")
print(x+y)
print("Subtraction")
print(x-y)
print("Multiplication")
print(x*y)
print("Division")
print(x/y)
```

**Another way to do multiplication operation in an array**

```python
print("Addition")
print(np.add(x, y))
print("Subtraction")
print(np.subtract(x, y))
print("Multiplication")
print(np.multiply(x, y))
print("Division")
print(np.divide(x, y))
```

**Two ways to find Transpose of an Array**

```python
arr = np.random.randint(10, 30, (2, 3))
print(arr)
print(arr.shape)
```

```python
# FIRST WAY
arr2 = np.transpose(arr)
print(arr2)
print(arr2.shape)
```

```python
# SECOND WAY
arr2 = arr.T
print(arr2)
print(arr2.shape)
```

**RESHAPING OF AN ARRAY**

```python
 x = np.random.randint(10, 30, (2, 3))
print(x)
print(x.shape)
```

```python
y = x.reshape(3, 2)
print(y)
print(y.shape)
```

#Matplotlib Basics

```python
# Importing the pandas library
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the Data

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
# cose wave
plt.plot(x, z)
plt.show()
```

```python
# data labelling
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# construct parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.title('parabola')
plt.plot(x, y)
plt.show()
```

# Bar Graph

```python
plot = plt.figure()
ax = plot.add_axes([0, 0, 1, 1])
subjects = ['Maths', 'English', 'Science', 'Computer', 'Hindi']
students = [80, 70, 85, 90, 65]
ax.bar(subjects, students)
plt.ylabel('Percentage of students pass in subjects')
plt.xlabel('Subjects')
plt.show()
```

#Pie Chart

```python
plot = plt.figure()
ax = plot.add_axes([0, 0, 1, 1])
languages = ['English', 'Tamil', 'Bengali', 'Gujrati', 'Hindi', 'Punjabi']
people = [80, 60, 50, 45, 85, 65]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

#Scatter Plot

```python
plot = plt.figure()
ax = plot.add_axes([0, 0, 1, 1])
x = np.linspace(10, 20, 40)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x, y, color = 'r')
ax.scatter(x, z, color = 'b')
plt.show()
```

#3D Scatter Plot

```python
plot = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap="Greens")
plt.show()
```
